```python
import numpy as np

np.random.seed(42)
days = 30
products = ['Electronics', 'Clothing', 'Books', 'Home']

lambda_customers = 50
daily_customers = np.random.poisson(lambda_customers, days)
cust_mean = np.mean(daily_customers)
cust_max = np.max(daily_customers)
cust_total = np.sum(daily_customers)

cust_per_day = daily_customers
all_purchases = []
for day in range(days):
    day_purchases = np.random.normal(75, 25, cust_per_day[day])
    day_purchases = np.maximum(day_purchases, 5)
    all_purchases.extend(day_purchases)
purchase_mean = np.mean(all_purchases)
purchase_max = np.max(all_purchases)
total_revenue = np.sum(all_purchases)

probs = [0.4, 0.3, 0.2, 0.1]
total_orders = len(all_purchases)
category_selections = np.random.multinomial(total_orders, probs)
category_pct = 100 * category_selections / total_orders

delivery_times = np.random.poisson(3, total_orders)
delivery_mean = np.mean(delivery_times)
delivery_max = np.max(delivery_times)
on_time_3days = np.sum(delivery_times <= 3) / total_orders * 100

scale_wait = 15
wait_times = np.random.exponential(scale_wait, total_orders-1)
wait_mean = np.mean(wait_times)
wait_p95 = np.percentile(wait_times, 95)

quality_results = np.random.binomial(1, 0.02, total_orders)
defects = np.sum(quality_results)
defect_rate = 100 * defects / total_orders
returns = defects * purchase_mean
net_revenue = total_revenue - returns

day_data = []
purchase_idx = 0
for day in range(days):
    day_rev = np.sum(all_purchases[purchase_idx:purchase_idx+daily_customers[day]])
    day_data.append([day+1, daily_customers[day], day_rev])
    purchase_idx += daily_customers[day]
day_df = np.array(day_data)
top_days = np.argsort(day_df[:, 2])[::-1][:5]

print("=== E-COMMERCE 30-DAY RESULTS ===")
print(f"Total Customers:     {cust_total:,}")
print(f"Gross Revenue:       ${total_revenue:,.0f}")
print(f"Returns:             ${returns:,.0f}")
print(f"Net Revenue:         ${net_revenue:,.0f}")
print(f"Orders:              {total_orders:,}")
print(f"Avg Delivery:        {delivery_mean:.1f} days")
print(f"Defect Rate:         {defect_rate:.2f}%")
print(f"Peak Customers:      {cust_max}")
print()
print("PRODUCT BREAKDOWN:")
for i, cat in enumerate(products):
    print(f"{cat:10s}: {category_selections[i]:4,} ({category_pct[i]:5.1f}%)")
print()
print("TOP 5 REVENUE DAYS:")
for i, day_idx in enumerate(top_days):
    day_num = int(day_df[day_idx, 0])
    custs = int(day_df[day_idx, 1])
    rev = day_df[day_idx, 2]
    print(f"Day {day_num:2d}: {custs:3,} cust, ${rev:8,.0f}")
```
